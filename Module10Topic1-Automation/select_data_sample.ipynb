```python
count = 0
for i, df in enumerate(pd.read_csv('../Data/2010HMDA.csv.gz', compression='gzip', chunksize=1_000_000, low_memory=False)):
    count += len(df)
    df.to_csv(f'../Data/2010HMDA/2010HMDA_{i+1}.csv.gz',
              index=False,
              compression='gzip')
print(f'Total rows: {count}')
del df
```
